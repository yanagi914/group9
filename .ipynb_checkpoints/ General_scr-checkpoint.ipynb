```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument('--headless')

browser = webdriver.Chrome(options=options)
url = 'https://u.muroran-it.ac.jp/kyomu/syllabustop2021/F6_.html'
browser.get(url)
```

```python
foms = []
a_front = '/html/body/table[2]/tbody/tr['
sum = 1
a_end = ']/td[5]/a'

for i in range(223):
    path = a_front + str(sum) + a_end
    browser.find_element_by_xpath(path).click()
    fom = browser.find_elements_by_xpath('//*[@id="tabs-1"]/table/tbody/tr[7]/td')[0].text
    foms.append(fom)
    browser.back()
    sum += 1
    sleep(0.5)

foms
```

```python
tans = []
a_front = '/html/body/table[2]/tbody/tr['
sum = 1
a_end = ']/td[5]/a'

for i in range(223):
    path = a_front + str(sum) + a_end
    browser.find_element_by_xpath(path).click()
    tan = browser.find_elements_by_xpath('//*[@id="tabs-1"]/table/tbody/tr[10]/td')[0].text
    tans.append(tan)
    browser.back()
    sum += 1
    sleep(0.5)

tans
```

```python
tans2 = []
for item in tans:
    
    #文字列の置換
    item_mod = item.replace(" ", "")
    
    #リストに追加
    tans2.append(item_mod)
```

```python
tans2
```

```python
codes = []
a_front = '/html/body/table[2]/tbody/tr['
sum = 1
a_end = ']/td[5]/a'

for i in range(223):
    path = a_front + str(sum) + a_end
    browser.find_element_by_xpath(path).click()
    code = browser.find_elements_by_xpath('//*[@id="tabs-1"]/table/tbody/tr[12]/td')[0].text
    codes.append(code)
    browser.back()
    sum += 1
    sleep(0.5)

codes
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['必修・選択'] = foms
df['単位数'] = tans2
df['時間割コード'] = codes
df
```

```python
df.to_csv('一般教養科目.csv',index=False)
```
